```python
import matplotlib.pyplot as plt
def segment_characters(image_path):
  
    
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    plt.imshow(image)  # Convert from BGR to RGB
    plt.axis('off')  # Hide axis
    plt.show()

    # Ensure that the loaded image is in grayscale format
    if len(image.shape) == 2:
        print("Image is already in grayscale format.")
        gray = image
    else:
        # Convert the loaded image to grayscale
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    
    plt.imshow(gray, cmap='gray')  # Convert from BGR to RGB
    plt.axis('off')  # Hide axis
    plt.show()
    
    

    # Apply thresholding to extract text regions
    #_, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 2)


    # Find contours in the thresholded image
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)



    # Iterate through contours to extract character regions
    character_images = []
    big_image=[]
    accepted_from_big=[]
    for contour in contours:
        # Get bounding box coordinates of the contour
        x, y, w, h = cv2.boundingRect(contour)
        
        if w >= 15 and h >= 15:
        # If dimensions match, add the contour to the list of filtered contours

            # Extract character region from the original image
            imag = image[y:y+h, x:x+w]
            _, segmented_region = cv2.threshold(imag, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
            plt.imshow(imag, cmap='gray')
            plt.axis('off')
            plt.show()
            
            cv2.imshow("Segmented Character", segmented_region)
            image_height, image_width = image.shape
            print(image_height, image_width, image_channels)
            
            
            if (image_height> 27 and image_width>17):
                
                window_size = (19,27)

                # Define step size for sliding window
                step_size = 2
                edge_densities = []
                x_coordinates = []
                y_coordinates = []

                # Iterate over the image using sliding window
                for y in range(0, imag.shape[0] - window_size[1] + 1, step_size):
                    for x in range(0, imag.shape[1] - window_size[0] + 1, step_size):
                        # Extract sub-image using sliding window
                        sub_image = imag[y:y+window_size[1], x:x+window_size[0]]

                        # Apply edge detection or any other analysis to determine if the sub-image should be discarded
                        # For example, you can use Canny edge detection to detect edges
                        edges = cv2.Canny(sub_image, 100, 200)

                        # Check if the sub-image contains a lot of edges
                        edge_density = cv2.countNonZero(edges) / (window_size[0] * window_size[1])

                        # If the edge density is below a certain threshold, discard the sub-image
                        if (edge_density > 0.19):  # Adjust the threshold value as needed
                            # Perform further processing with the sub-image
                            print("Sub-image at position ({}, {}) passed edge density test.".format(x, y))
                            print(edge_density)
                            edge_densities.append(edge_density)
                            x_coordinates.append(x)
                            y_coordinates.append(y)
                            plt.imshow(sub_image, cmap='gray')
                            plt.axis('off')
                            plt.show()
                            big_image.append(sub_image)
            

            # Add character region to the list
            character_images.append(image)
            #top_10_indices = sorted(range(len(edge_densities)), key=lambda i: edge_densities[i], reverse=True)[:20]


    check_x, check_y= get_from_big(big_image,3,x_coordinates, y_coordinates, edge_densities)
    for i in range(0, len(check_x)):
        sub_image = imag[check_y[i]:check_y[i]+window_size[1], check_x[i]:check_x[i]+window_size[0]]
        plt.imshow(sub_image, cmap='gray')
        plt.axis('off')
        plt.show()
    return character_images
```
